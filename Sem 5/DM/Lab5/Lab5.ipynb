### 2)  Handle Missing Values in data set [use dropna(), fillna(), and interpolate]


```python
data_dropped_any = data.dropna(how='any',axis=0)
data_dropped_any
data_dropped_all = data.dropna(how='all',axis=0)
data_dropped_all
```

```python
# data.fillna('-')

filled_data = data.fillna({'Age': int(data.Age.mean()), 'Cabin': data.Cabin.mode()[0] })
filled_data
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
data_copy = data.copy()
data_copy.Age = data_copy.Age.interpolate()
age = data_copy.Age
age_min = age.min()
age_max = age.max()
new_min = 0
new_max = 1

data_copy['Age_Min_Max_Scale'] = ((age - age_min)/(age_max - age_min))*((new_max - new_min)+new_min)

data_copy[['Age','Age_Min_Max_Scale']].head()
```

```python
max_abs_age = data_copy['Age'].abs().max()
j = len(str(int(max_abs_age)))
data_copy['Age_decimal_scaled'] = data_copy['Age'] / (10 ** j)
data_copy[['Age', 'Age_decimal_scaled']].head()
```

```python
age_mean = age.mean()
age_sd = age.std()
data_copy['Age_ZScore'] = (age - age_mean)/age_sd
data_copy[['Age', 'Age_ZScore']].head()
```
